# Gradient Descent

Notebook imports packages

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
%matplotlib inline
```

### Example 1

#### $$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2+x+1
```

Make Data

```python
x_1 = np.linspace(start = -3, stop = 3, num = 500)
```

Now all we got to do is Plot the graph

```python
plt.xlim(-3, 3)
plt.ylim(-1, 12)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(X)', fontsize=16)
plt.plot(x_1, f(x_1))
plt.show()
```

### Finding Derivative of the function

```python
def df(x):
    return 2*x+1
```

```python
plt.figure(figsize=[15,5])

plt.subplot(1,2,1)
plt.title('Cost Function', fontsize=17)
plt.xlim(-3, 3)
plt.ylim(-1, 12)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(X)', fontsize=16)
plt.plot(x_1, f(x_1), color='blue', )

plt.subplot(1,2,2)
plt.title("Slope of Cost Function", fontsize=17)
plt.xlim(-3,3)
plt.ylim(-5,7)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(X)', fontsize=16)
plt.grid()
plt.plot(x_1,df(x_1), color='green')


plt.show()
```

### Gradient Descent

```python
new_x = 3
previous_x = 0
step_multiplier = 0.1
count = 0
precision = 0.001

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier*gradient
    step_size = abs(new_x-previous_x)
    if step_size < precision:
        break
    else:
        count+=1
        
        
print('Local Minima at:', new_x)
print('Slope or df(x) value:',df(new_x))
print('Cost at this point:', f(new_x))
print('Count:',count)
```
